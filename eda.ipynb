```
# Import Required Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

```
# Load the Dataset
data = pd.read_csv('dummy_data.csv')

```

```
# Check for Null/NaN Values
print(data.isnull().sum())

```

```
# Plot a Histogram for a Feature
plt.hist(data['Age'])
plt.title('Age Distribution')
plt.show()

```

```
# Visualize a Correlation Matrix
corr_matrix = data.corr()
sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Matrix')
plt.show()

```

```
# Boxplot for Outliers
sns.boxplot(data=data,x='TicketClass',y='Age')
plt.title('Boxplot for Outliers Detection')
plt.show()

```

```
# Distribution Plot
sns.kdeplot(data=data['Age'], shade=True)
plt.title('Distribution Plot')
plt.show()

```

```
# Pairplot for Multiple Variables
sns.pairplot(data=data, hue='Sex')
plt.title('Pairplot Visualizing Multiple Variables')
plt.show()

```
